```python
#Importing Things
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

#reading in the data set
df = pd.read_csv('user-data.csv')
df = df.loc[:,'hits':'ranking']
```

```python
#looking at the data set. It looks like weapontype, gametype should be switched to dummy variables. Hits through deaths
#should be standardized.
df.head()
```

```python
#making the dummy variables
dummy = pd.get_dummies(df.loc[:,['weapontype','gametype']])
df = df.drop(['weapontype','gametype'], axis=1)
df = pd.concat([df,dummy],axis=1)
df.head()
```

```python
#seperating in between training and testing data sets
X = df.drop(['kdRatio'],axis=1)
y = df.loc[:,'kdRatio']
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.25,random_state=246)
```

```python
#scaling the data set
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train.loc[:,'hits':'deaths'])
names = X_train.loc[:,'hits':'deaths'].columns
scaled_X_train = pd.concat([X_train.loc[:,'weapon':'gametype_xbl'],pd.DataFrame(scaled_X_train,columns=names)],axis=1)
scaled_X_test = scaler.transform(X_test.loc[:,'hits':'deaths'])
scaled_X_test = pd.concat([X_test.loc[:,'weapon':'gametype_xbl'],pd.DataFrame(scaled_X_test,columns=names)],axis=1)
```

```python
#Writing out the data sets for the model
scaled_X_train.to_csv('scaled_X_train.csv')
scaled_X_test.to_csv('scaled_X_test.csv')
X_train.to_csv('X_train.csv')
X_test.to_csv('X_test.csv')
y_train.to_csv('y_train.csv')
y_test.to_csv('y_test.csv')
```
